#fizzBuzz

```python
#fizzbuzz
def fizzBuzz(n):
  for i in range(1,n):
    if(i%5==0 and i%3 == 0):
      print(str(i)+" = FizzBuzz")
    elif(i%3==0):
      print(str(i)+" = Fizz")
    elif(i%5==0):
      print(str(i)+" = Buzz")
    else:
      print(i)

fizzBuzz(15)
```

#Permutation Jumble words

```python
import random
def choose():
  word = ['rainbow','computer','science','programming','mathematics','player','draw','permutation','resume','date']
  return random.choice(word)

def jumble(word):
  return "".join(random.sample(word,len(word))) # sample chooses k unique element from given sequence

def thank(p1name,p2name,p1points,p2points):
  print(p1name,"Your score is: ",p1points)
  print(p2name,"Your score is: ",p2points)
  print("------------x------------")
  if p1points == p2points:
    print("it's draw")
  elif p1points > p2points:
    print("Congratulation",p1name)
  else:
    print("Congratulation",p2name)

def play():
  p1name = input("Player 1, Enter your name:")
  p2name = input("Player 2, Enter your name:")
  p1Points = 0
  p2Points = 0
  turn = 0
  while(1):
    #computer's task
    picked_word = choose()
    #create question
    ques = jumble(picked_word)
    print(ques)
    # player 1
    if(turn == 0):
      print(p1name,", Your Turn,")
      ans = input("what is on my mind? ")
      if ans ==picked_word:
        print("Hurray right guess")
        p1Points += 1
        print("Your score is",p1Points)
      else:
        print("Better luck next time, Correct answer is",picked_word)
      turn = 1
      c = int(input('Press 0 to quit: '))
      if c == 0:
        thank(p1name,p2name,p1Points,p2Points)
        break

    # player 2
    else:
      print(p2name,", Your Turn,")
      ans = input("what is on my mind")
      if ans ==picked_word:
        print("Hurray right guess")
        p2Points += 1
        print("Your score is",p2Points)
      else:
        print("Better luck next time, Correct answer is",picked_word)
      turn = 0
      c = int(input('Press 1 to continues and 0 to quit: '))
      if c == 0:
        thank(p1name,p2name,p1Points,p2Points)
        break
    print("------------x------------")

play()
```

#Programming Assignments

Create a Python program that finds the second largest number in a list of positive integers(includes zero). The program should prompt the user to input a list of numbers, then compute and print the second largest number in that list.

Input Format:
The input consists of a single list of numbers, separated by spaces.
Hint: Use .split() function to convert input to list.

Output Format:
The output consists of the second largest number in the input list.

Example:

Input:
3 1 4 1 5 9 2 6 5 3 5

Output:
6

```python
input_numbers = input()

numbers_list = input_numbers.split()
for i in range(len(numbers_list)):
    numbers_list[i] = int(numbers_list[i])

first_largest = -1
second_largest = -1

for number in numbers_list:
    if number > first_largest:
        second_largest = first_largest
        first_largest = number
    elif number > second_largest and number != first_largest:
        second_largest = number

print(second_largest)
```

Create a Python program that removes all duplicate positive integer numbers (includes zero) from a list and prints the unique numbers in the order they first appeared.
The program should prompt the user to input a list of numbers, then process the list to remove duplicates and print the resulting list of unique numbers.

Input Format:
The input consists of a single list of numbers, separated by spaces.

Output Format:
The output consists of the unique numbers, separated by spaces, from the input list, in the order they first appeared.

Example:

Input:
3 1 4 1 5 9 2 6 5 3 5

Output:
3 1 4 5 9 2 6

```python
input_numbers = input()

numbers_list = input_numbers.split()
for i in range(len(numbers_list)):
    numbers_list[i] = int(numbers_list[i])

unique_numbers = []

for number in numbers_list:
    if number not in unique_numbers:
        unique_numbers.append(number)

for number in unique_numbers:
    print(number, end=" ")
```

Create a Python program that takes a list of integers, reverses the list, adds the values at odd indices from both the original and reversed lists, and creates a new list with the result. The new list should be printed in the end.

Input Format:
The input consists of a single list of integers, separated by spaces.

Output Format:
The output consists of the new list of values, separated by spaces, obtained by adding values at odd indices from both the original and reversed lists.


Example:
Input:
1 2 3 4 5

Output:
1 6 3 6 5

```python
# Prompting user for input
input_numbers = input()

# Converting input string to a list of integers
numbers_list = input_numbers.split()
for i in range(len(numbers_list)):
    numbers_list[i] = int(numbers_list[i])

# Reversing the list
reversed_list = numbers_list[::-1]

# Creating the new list by adding values at odd indices
new_list = []
for i in range(len(numbers_list)):
    if i % 2 == 0:
        # For even indices (0-based index), just append the original value
        new_list.append(numbers_list[i])
    else:
        # For odd indices, add values from both original and reversed lists
        new_value = numbers_list[i] + reversed_list[i]
        new_list.append(new_value)

# Printing the new list
for i in new_list:
    print(i, end=" ")
```
