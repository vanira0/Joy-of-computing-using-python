# Permutation Jumble words

````python
import random
def choose():
  word = ['rainbow','computer','science','programming','mathematics','player','draw','permutation','resume','date']
  return random.choice(word)

def jumble(word):
  return "".join(random.sample(word,len(word))) # sample chooses k unique element from given sequence

def thank(p1name,p2name,p1points,p2points):
  print(p1name,"Your score is: ",p1points)
  print(p2name,"Your score is: ",p2points)
  print("------------x------------")
  if p1points == p2points:
    print("it's draw")
  elif p1points > p2points:
    print("Congratulation",p1name)
  else:
    print("Congratulation",p2name)

def play():
  p1name = input("Player 1, Enter your name:")
  p2name = input("Player 2, Enter your name:")
  p1Points = 0
  p2Points = 0
  turn = 0
  while(1):
    #computer's task
    picked_word = choose()
    #create question
    ques = jumble(picked_word)
    print(ques)
    # player 1
    if(turn == 0):
      print(p1name,", Your Turn,")
      ans = input("what is on my mind? ")
      if ans ==picked_word:
        print("Hurray right guess")
        p1Points += 1
        print("Your score is",p1Points)
      else:
        print("Better luck next time, Correct answer is",picked_word)
      turn = 1
      c = int(input('Press 0 to quit: '))
      if c == 0:
        thank(p1name,p2name,p1Points,p2Points)
        break

    # player 2
    else:
      print(p2name,", Your Turn,")
      ans = input("what is on my mind")
      if ans ==picked_word:
        print("Hurray right guess")
        p2Points += 1
        print("Your score is",p2Points)
      else:
        print("Better luck next time, Correct answer is",picked_word)
      turn = 0
      c = int(input('Press 1 to continues and 0 to quit: '))
      if c == 0:
        thank(p1name,p2name,p1Points,p2Points)
        break
    print("------------x------------")

play()
````

#Magic square: Hit and trial


**Fact:-**

*M = n(n^2+1)/2*

**Steps:-**

In any magic square,
1. 1 is located at position (n/2, n-1).

2.   Let's say the position of 1 i.e. (n/2, n-1) is (p, q), then next number which is 2 is located at (p-1, q+1) position. Anytime if the calculated row position becomes -1 then make it n-1 and if column position becomes n then make it 0.
3.   If the calculated position already contains a number then decrement the column by 2 and increment the row by 1.
4.   If anytime row position becomes -1 and column becomes n, switch it to (0, n-2).

````python
from ast import Continue
def magic_square(n):
  #Initialize magic square with zeros
  magicSquare = [[0]*n for _ in range(n)]

  # Initial position
  i = n//2
  j = n-1

  num = n*n
  count = 1

  while count<=num:
    # condition 4: If both row and column are out of bounds
    if(i == -1 and j == n):
      j = n-2
      i = 0
    else:
      if(j == n):     #column value is exceeding
        j = 0;
      if(i<0):        #row is becoming -1
        i = n-1

    # condition 2: if current cell is already filled
    if magicSquare[i][j] != 0:
      j -= 2
      i += 1
      continue
    else:
      # place the current Number
      magicSquare[i][j] = count
      count += 1
    # Move to next position
    i -= 1
    j += 1


  sum = n*((n**2)+1)//2
  print("Sum of each row/col/diagnol: "+str(sum))
  # print magic square
  for row in magicSquare:
    print(" ".join(str(x) for x in row))


n = int(input())
magic_square(n)
````

````python
import random
import datetime

def checkLeapYear(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

birthday = []
i = 0
while i < 50:
    year = random.randint(1902, 2024)
    month = random.randint(1, 12)

    if month == 2:  # February
        if checkLeapYear(year):
            day = random.randint(1, 29)
        else:
            day = random.randint(1, 28)
    elif month in [4, 6, 9, 11]:  # April, June, September, November
        day = random.randint(1, 30)
    else:  # January, March, May, July, August, October, December
        day = random.randint(1, 31)

    dd = datetime.date(year, month, day)
    day_of_year = dd.timetuple().tm_yday
    birthday.append(day_of_year)
    i += 1

birthday.sort()

i = 0
while i < 50:
    print(birthday[i])
    i += 1
````

````python
def binary_search(a,x):
  s = 0
  e = len(a)-1
  cnt=0
  while(s<e):
    cnt+=1
    mid = s + (e-s)//2
    if(x == a[mid]):
      return mid
    elif(x>a[mid]):
      s = mid+1
    else:
      e = mid-1
  return -1
a = [1,2,3,4,5,5]

binary_search(a,6)
````

````python
import numpy as np
board = np.full((3,3),'-')
p1s = 'X'
p2s = 'O'

def check_row(symbol):
  for r in range(3):
    count = 0
    for c in range(3):
      if board[r][c] == symbol:
        count += 1;
    if count == 3:
      print(symbol, 'won')
      return True
  return False

def check_col(symbol):
  for c in range(3):
    count = 0
    for r in range(3):
      if board[r][c] == symbol:
        count += 1;
    if count == 3:
      print(symbol, 'won')
      return True
  return False

def check_diagonal(symbol):
  if board[0][2] == board[1][1] and board[1][1] == board[2][0] and board[1][1] == symbol:
    print(symbol, 'won')
    return True
  if board[0][0] == board[1][1] and board[1][1] == board[2][2] and board[1][1] == symbol:
    print(symbol, 'won')
    return True
  return False

def won(symbol):
  return check_row(symbol) or check_col(symbol) or check_diagonal(symbol)

def place(symbol):
  print(np.matrix(board))
  while(1):
    row = int(input('Enter row (1/2/3): '))
    col = int(input('Enter col (1/2/3): '))
    if row>0 and row<4 and col>0 and col<4 and board[row-1][col-1] == '-':
      break;
    else:
      print("Invalid input. Enter again")
  board[row-1][col-1] = symbol

def play():
  for turn in range(9):
    if(turn%2 == 0):
      print("X turn")
      place(p1s)
      if won(p1s):
        break
    else:
      print("O turn")
      place(p2s)
      if won(p2s):
        break
  if not(won(p1s)) and not(won(p2s)):
    print('Draw')

play()
````

#Snake and Ladder

````python
import random
from PIL import Image
end = 100
def show_board():
  img = Image.open('slb.jpg')
  img.show()

def roll_die():
  return random.randint(1,6)

def check_ladder(points):
  ladder = {8:26,21:82,43:77,50:91,54:93,62:96, 66:87,80:100}
  return ladder[points] if points in ladder else points

def check_snake(points):
  snake = {44:22,46:5,48:9,52:11,55:7,59:17,64:36,73:1,83:19,92:51,95:24,98:28}
  return snake[points] if points in snake else points

def reached_end(points):
  if points == end:
    return True
  else:
    return False

def play():
  p1_name = input('Player 1 name: ')
  p2_name = input('Player 2 name: ')
  pp1 = 0
  pp2 = 0
  turn = 0
  while(1):
    if turn%2 == 0:
      # player 1 turn
      print(p1_name, 'your turn')
      c = input("Do want to continue? n/y ")
      if c == 'n':
        print(p1_name,' scored',pp1)
        print(p2_name,' scored',pp2)
        print(p1_name, ' Quit the game')
        break;
      dice = roll_die()
      print('Dice show: ', dice)
      pp1 += dice
      pp1 = check_ladder(pp1)
      pp1 = check_snake(pp1)
      if pp1 > end:
        pp1 -= dice
      print(p1_name, 'Your score: ',pp1)
      if reached_end(pp1):
        print(p1_name,' won')
        break
    else:
      #player 2 turn
      print(p2_name, 'your turn')
      c = input("Do want to continue? n/y ")
      if c == 'n':
        print(p1_name,' scored',pp1)
        print(p2_name,' scored',pp2)
        print(p2_name, ' Quit the game')
        break;
      dice = roll_die()
      print('Dice show: ', dice)
      pp2 += dice
      pp2 = check_ladder(pp2)
      pp2 = check_snake(pp2)
      if pp2 > end:
        pp2 -= dice
      print(p2_name, 'Your score: ',pp2)
      if reached_end(pp2):
        print(p2_name,' won')
        break
    turn += 1


show_board()
play()
````

````python
import networkx as nx
import numpy as np

# Load the graph
G = nx.read_edgelist('facebook_combined.txt')

# Get the all-pairs shortest path lengths
all_spl_lengths = dict(nx.all_pairs_shortest_path_length(G))

# Initialize variables to compute min, max, and sum for average
min_spl = float('inf')
max_spl = float('-inf')
spl_sum = 0
spl_count = 0

# Iterate through shortest path lengths
for u, spl_dict in all_spl_lengths.items():
    for v, length in spl_dict.items():
        if u != v:
            spl_sum += length
            spl_count += 1
            if length < min_spl:
                min_spl = length
            if length > max_spl:
                max_spl = length

# Compute the average shortest path length
avg_spl = spl_sum / spl_count if spl_count > 0 else 0

# Print the results
print("Minimum shortest path length:", min_spl)
print("Maximum shortest path length:", max_spl)
print("Average shortest path length:", avg_spl)
print(all_spl_lengths)
````

````python
import networkx as nx
import random
import matplotlib.pyplot as plt

def add_edges():
  nodes= list(G.nodes())
  for s in nodes:
    for t in nodes:
      if s != t:
        r = random.random()
        if r<=0.5:
          G.add_edge(s,t)
  return G

# create a directed graph
G = nx.DiGraph()
G.add_nodes_from([i for i in range(10)])
G = add_edges()

# visualize the graph
nx.draw(G,with_labels = True)
plt.show()
````

````python

def assign_points(G):
  nodes = list(G.nodes())
  p = []
  for each in nodes:
    p.append(100)
  return p

def distribute_points(G,points):
  nodes = list(G.nodes())
  new_points = []
  for i in range(len(nodes)):
    new_points.append(0)
  for n in nodes:
    out = list(G.out_edges(n))
    if len(out) == 0:
      new_points[n] = new_points[n] + points[n]
    else:
      share = points[n]/len(out)
      for (src,tgt) in out:
        new_points[tgt] = new_points[tgt] + share
  return new_points

def keep_distributing(points, G):
  while(1):
    new_points = distribute_points(G,points)
    print(new_points)
    points = new_points
    stop = input("Press # to stop or any other key to continue")
    if(stop == "#"):
      break;
  return new_points

def rank_by_points(points):
  d = {}
  for i in range(len(points)):
    d[i] = points[i]
  print(sorted(d.items(),key = lambda f:f[1]))



#Assign initial points
points = assign_points(G)

# keep distributing
final_points = keep_distributing(points,G)

# rank by points
rank_by_points(final_points)
````

````python
from itertools import permutations

def number_game(input1, input2, input3, input4):
    D = input2
    E = input3
    F = input4

    # Generate all permutations of F and convert them to a list of strings
    possible_numbers = sorted(set("".join(p) for p in permutations(F) if "".join(p) != F))

    for X in possible_numbers:
        if X > D and X > E:
            if X[::-1] < D[::-1] and X[::-1] < E[::-1]:
                return X

    return "-1"
print(number_game(3, "124", "324", "335"))
````

````python
from collections import deque

def min_moves_to_cross(N, A):
    # Initialize visited array
    visited = [False] * (N + 1)

    # Create a dictionary to map values to their positions
    value_positions = {}
    for i, val in enumerate(A):
        if val not in value_positions:
            value_positions[val] = []
        value_positions[val].append(i+1)

    # Queue for BFS, storing position and moves
    queue = deque([(1, 0)])  # Start from the first brick with 0 moves

    while queue:
        current_position, moves = queue.popleft()

        # If reached the target, return the moves
        if current_position == N:
            return moves

        # Check next brick
        if current_position < N:
            queue.append((current_position + 1, moves + 1))

        # Check previous brick
        if current_position > 1:
            queue.append((current_position - 1, moves + 1))

        # Check bricks with the same value
        if A[current_position - 1] in value_positions:
            for pos in value_positions[A[current_position - 1]]:
                if not visited[pos]:
                    visited[pos] = True
                    queue.append((pos, moves + 1))

    return -1  # Should never reach here if input is valid

# Test cases
print(min_moves_to_cross(6, [5,2,3,4,5,3]))  # Expected output: 2
print(min_moves_to_cross(5, [1,2,3,4,1]))   # Expected output: 1
````

````python
def getOutlierValue(arr):
    total_sum = sum(arr)
    max_outlier = float('-inf')

    for num in arr:
        remaining_sum = total_sum - num

        # Count occurrences of the current number
        count = arr.count(num)

        # Check if the current number appears less than twice
        if count < 2:
            # Check if it's not equal to the sum of the remaining numbers
            if remaining_sum != num:
                # Update max_outlier if current number is larger
                max_outlier = max(max_outlier, num)

    return max_outlier
# For input 5,9,2,8,1 ,5
# Output should be 9 but itscoming5
# Example usage
arr = [5,9,2,8,1,5]
print("Greatest outlier value:", getOutlierValue(arr))
````

````python
def findPairs(cost, x):
    remainder_counts = {}
    total_pairs = 0

    for c in cost:
        complement = x - (c % x)
        if complement in remainder_counts:
            total_pairs += remainder_counts[complement]

        remainder_counts[c % x] = remainder_counts.get(c % x, 0) + 1

    # Handle case where x divides evenly into the sum of two identical numbers
    if 0 in remainder_counts:
        total_pairs += remainder_counts[0] * (remainder_counts[0] - 1) // 2

    return total_pairs

# Example usage
cost = [31, 25, 85, 29, 35]
x = 60
result = findPairs(cost, x)
print(f"Number of pairs: {result}")
````

````python
def min_ascii_distance(S, A):
    # Convert strings to sets for efficient lookup
    set_S = set(S)
    set_A = set(A)

    # Initialize variables
    total_distance = 0
    chars_not_in_S = 0

    # Iterate through each unique character in A
    for char in set_A:
        if char not in set_S:
            chars_not_in_S += 1

            # Find the closest character in S
            closest_char = min(set_S, key=lambda x: abs(ord(x) - ord(char)))

            # Calculate the ASCII distance
            distance = abs(ord(closest_char) - ord(char))

            # Add the distance to the total
            total_distance += distance

    # Return 0 if no modifications were needed, otherwise return the total distance
    return total_distance if chars_not_in_S > 0 else 0

# Test cases
print(min_ascii_distance("bfjpy", "abcd"))  # Output: 4
print(min_ascii_distance("abcdefgh", "aghfedbc"))  # Output: 0
````
